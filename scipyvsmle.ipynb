### Fit PM to curve using 5e6thresholdpm.txt data

```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
```

```python
#d = 3
def pseudeothreshold_function(x, a, b):
   return a * np.power(3*x, 3*b) 
popt, pcov = curve_fit(pseudeothreshold_function, np.array(phys_lists_thresholdpm[0]), np.array(log_lists_thresholdpm[0]))
a_optimal, b_optimal = popt
y_fit = pseudeothreshold_function(np.array(phys_lists_thresholdpm[0]), a_optimal, b_optimal)
plt.scatter(np.array(phys_lists_thresholdpm[0]), np.array(log_lists_thresholdpm[0]), label="Data")
plt.plot(np.array(phys_lists_thresholdpm[0]), y_fit, label="Fitted Curve", color='red')
plt.title("fit d = 3 for PM")
plt.legend()
plt.show()
print(a_optimal)
```

```python
#d = 5
def pseudeothreshold_function(x, a, b):
   return a * np.power(5*x, 5*b) 
popt, pcov = curve_fit(pseudeothreshold_function, np.array(phys_lists_thresholdpm[1]), np.array(log_lists_thresholdpm[1]))
a_optimal, b_optimal = popt
y_fit = pseudeothreshold_function(np.array(phys_lists_thresholdpm[1]), a_optimal, b_optimal)
plt.scatter(np.array(phys_lists_thresholdpm[1]), np.array(log_lists_thresholdpm[1]), label="Data")
plt.plot(np.array(phys_lists_thresholdpm[1]), y_fit, label="Fitted Curve", color='red')
plt.title("fit d = 5 for PM")
plt.legend()
plt.show()
print(a_optimal)
```

```python
count = 3
for d in range(3,16,2):
    def pseudeothreshold_function(x, a, b):
       return a * np.power(d*x, d*b) 
    popt, pcov = curve_fit(pseudeothreshold_function, np.array(phys_lists_thresholdpm[d-count]), np.array(log_lists_thresholdpm[d-count]))
    a_optimal, b_optimal = popt
    y_fit = pseudeothreshold_function(np.array(phys_lists_thresholdpm[d-count]), a_optimal, b_optimal)
    plt.scatter(np.array(phys_lists_thresholdpm[d-count]), np.array(log_lists_thresholdpm[d-count]), label="Data")
    plt.plot(np.array(phys_lists_thresholdpm[d-count]), y_fit, label="Fitted Curve", color='red')
    plt.title("fit d="+str(d)+" for PM")
    plt.legend()
    plt.show()
    print(a_optimal)
    count += 1
```
