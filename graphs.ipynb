```python
import matplotlib.pyplot as plt
```

```python
def prob_d(p,w,d):
    return (2**w)*((p)**((w+1)/2))

def log_prob(p,list_w,list_comb, d):
    sum = 0
    for i in range(len(list_w)):
        sum += prob_d(p,list_w[i], d) * list_comb[i] 
    return sum
def plot(d):
    phys = [round(i * 0.01, 2) for i in range(51)]
    list_w = list(range(d, d**2, 2))
    if d == 3:
        list_comb = [6,9,0,1]
    if d == 5:
        list_comb = [120,2200,7825,17000,20700,12840,4150,600,100,0,1]
    log = []
    for p in phys:
        log.append(log_prob(p,list_w, list_comb, d))
    # print(phys)
    fig,ax = plt.subplots(1,1,dpi=200,figsize = (4,3))
    ax.plot(phys,log)
    ax.legend(fontsize=6)
    ax.set_xlabel("Physical Error Rate (p)")
    ax.set_ylabel("Logical Error Rate")
    plt.plot()


    

```

```python
plot(3)
```

```python
plot(5)
```

```python

phys = [round(i * 0.01, 2) for i in range(101)]
list_w_three = list(range(3, 9, 2))
list_w_five = list(range(5, 25, 2))
list_comb_three = [6,9,0,1]
list_comb_five = [120,2200,7825,17000,20700,12840,4150,600,100,0,1]
log_three= []
log_five= []
for p in phys:
    log_three.append(log_prob(p,list_w_three, list_comb_three, 3))
    log_five.append(log_prob(p,list_w_five, list_comb_five, 5))
# print(phys)
fig,ax = plt.subplots(1,1,dpi=200,figsize = (4,3))
ax.plot(phys,log_three)
ax.plot(phys,log_five)
ax.legend(fontsize=6)
ax.set_xlabel("Physical Error Rate (p)")
ax.set_ylabel("Logical Error Rate")
plt.plot()
```
